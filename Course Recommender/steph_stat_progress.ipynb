```python
#per student, count & identify

def list_courses_taken(student_id):#returns course id, input is a string (student's name)
    return [item.get_course_nr() for item in recommender.grade_data if item.name == student_id]

def count_courses_taken(student_id): #give student_name in string. duplicates not excluded. e.g  if someone took the same course more than once, or just once but recorded more than once, the duplicates are counted. dropped courses also not excluded 
    return len(list([item.get_course_nr() for item in recommender.grade_data if item.name == student_id]))

print list_courses_taken('6aef7c29444313899ca4610c1b0ad014d57bb5aa')
print count_courses_taken('6aef7c29444313899ca4610c1b0ad014d57bb5aa')
```

```python
#per student, return a set of grades for all course taken:
def list_student_grades(student_id):
    return [item.get_grade() for item in recommender.grade_data if item.name == student_id and item.get_grade() >= 0]

print list_student_grades('6aef7c29444313899ca4610c1b0ad014d57bb5aa')

len(list_student_grades('6aef7c29444313899ca4610c1b0ad014d57bb5aa')) == len(list_courses_taken('6aef7c29444313899ca4610c1b0ad014d57bb5aa')) == count_courses_taken('6aef7c29444313899ca4610c1b0ad014d57bb5aa')
```

```python
#per student, count & identify

def list_courses_passed(student_id):
    return [item.get_course_nr() for item in recommender.grade_data if item.name == student_id and item.get_grade() >= 6]

def count_courses_passed(student_id):
    return len([item.get_course_nr() for item in recommender.grade_data if item.name == student_id and item.get_grade() >= 6])

print list_courses_passed('6aef7c29444313899ca4610c1b0ad014d57bb5aa')
print count_courses_passed('6aef7c29444313899ca4610c1b0ad014d57bb5aa')
```

```python
def list_courses_failed(student_id):
 return [item.get_course_nr() for item in recommender.grade_data if item.name == student_id and item.get_grade() < 6]
    
def count_courses_failed(student_id):
    return len([item.get_course_nr() for item in recommender.grade_data if item.name == student_id and item.get_grade() < 6])

print list_courses_failed('6aef7c29444313899ca4610c1b0ad014d57bb5aa')
print count_courses_failed('6aef7c29444313899ca4610c1b0ad014d57bb5aa')

count_courses_passed('6aef7c29444313899ca4610c1b0ad014d57bb5aa') + count_courses_failed('6aef7c29444313899ca4610c1b0ad014d57bb5aa') == count_courses_taken('6aef7c29444313899ca4610c1b0ad014d57bb5aa') 
```

```python
#per student, return fail rate: (number of courses a student failed / total number of courses student took)

from decimal import Decimal

def per_student_fail_rate(student_id):
    return Decimal(count_courses_failed(student_id)) / Decimal(count_courses_taken(student_id))

print per_student_fail_rate('6aef7c29444313899ca4610c1b0ad014d57bb5aa')
```

```python
#per student, returan average grades:

import numpy as np

def per_student_avg_grade(student_id):
    return np.mean(list_student_grades(student_id))

print per_student_avg_grade('6aef7c29444313899ca4610c1b0ad014d57bb5aa')
```
